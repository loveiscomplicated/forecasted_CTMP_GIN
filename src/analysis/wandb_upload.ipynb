```python
import wandb
import json
import os

def upload_jsonl_to_wandb(file_path, project_name, run_name=None, config=None):
    """
    JSONL 파일의 로그 데이터를 WandB에 업로드합니다.
    
    Args:
        file_path (str): .jsonl 파일 경로
        project_name (str): WandB 프로젝트 이름
        run_name (str, optional): 실행(Run) 이름. 미지정 시 자동 생성.
        config (dict, optional): 모델 설정값 등 실험 메타데이터
    """
    if not os.path.exists(file_path):
        print(f"Error: 파일을 찾을 수 없습니다. ({file_path})")
        return

    # 1. WandB 초기화
    wandb.init(project=project_name, name=run_name, config=config)

    print(f"Uploading {file_path} to WandB project '{project_name}'...")

    # 2. 파일 읽기 및 로깅
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            
            # 'epoch'를 step으로 사용
            if "epoch" in data:
                step = data.pop("epoch")
                wandb.log(data, step=step)
            else:
                # epoch 정보가 없는 라인(예: 최종 test 결과)은 그냥 로깅
                wandb.log(data)

    # 3. 종료
    wandb.finish()
    print("Upload complete!")
```

```python
import numpy as np
from scipy import stats

# 1. 각 모델의 K-fold 최종 Test AUC 값 입력
ctmp_gin_auc = [[0.95539, 0.95488, 0.95517, 0.95493, 0.95506], [0.95365, 0.95374, 0.95369, 0.95379, 0.95378], [0.95386, 0.95431, 0.95346, 0.95443, 0.95389]]
gin_auc      = [[0.95339, 0.95356, 0.95344, 0.95351, 0.95322], [0.95336, 0.95325, 0.95343, 0.95312, 0.95321], [0.95361, 0.95321, 0.95367, 0.95318, 0.95368]]
a3tgcn_auc   = [0.95235, 0.95210, 0.95286, 0.95268, 0.95268, 0.95223, 0.95247, 0.95241, 0.95247, 0.95285, 0.95257, 0.95310, 0.95301, 0.95296, 0.95264]

def _mean_diff_ci(diff, alpha=0.05):
    """
    diff: array of (group1 - group2)
    returns: (mean_diff, (ci_low, ci_high))
    """
    diff = np.asarray(diff, dtype=float)
    n = diff.size
    mean_diff = diff.mean()
    sd = diff.std(ddof=1)
    se = sd / np.sqrt(n)
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
    ci_low = mean_diff - t_crit * se
    ci_high = mean_diff + t_crit * se
    return mean_diff, (ci_low, ci_high)

def _cohens_d_paired(diff):
    """Cohen's dz for paired samples."""
    diff = np.asarray(diff, dtype=float)
    return diff.mean() / diff.std(ddof=1)

def _cohens_d_welch(group1, group2):
    """Cohen's d for independent samples using pooled SD (classic)."""
    x = np.asarray(group1, dtype=float)
    y = np.asarray(group2, dtype=float)
    n1, n2 = x.size, y.size
    s1, s2 = x.std(ddof=1), y.std(ddof=1)
    sp = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2) / (n1+n2-2))
    return (x.mean() - y.mean()) / sp

def perform_t_test(group1, group2, baseline_name, test_type='paired', alpha=0.05):
    group1 = np.asarray(group1, dtype=float)
    group2 = np.asarray(group2, dtype=float)

    mean1, std1 = group1.mean(), group1.std(ddof=1)
    mean2, std2 = group2.mean(), group2.std(ddof=1)

    print(f"--- 모델 성능 요약 ({test_type} T-test) ---")
    print(f"CTMP_GIN      : Mean = {mean1:.5f}, Std = {std1:.5f}")
    print(f"{baseline_name:<12}: Mean = {mean2:.5f}, Std = {std2:.5f}")
    print("-" * 40)

    if test_type == 'paired':
        if group1.size != group2.size:
            raise ValueError("paired t-test는 두 그룹 길이가 같아야 합니다.")
        t_stat, p_value = stats.ttest_rel(group1, group2)

        diff = group1 - group2
        mean_diff, (ci_low, ci_high) = _mean_diff_ci(diff, alpha=alpha)
        d = _cohens_d_paired(diff)  # dz

        print(f"Mean(CTMP_GIN - {baseline_name}) = {mean_diff:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{ci_low:.6f}, {ci_high:.6f}]")
        # print(f"Cohen's dz (paired effect size) = {d:.3f}")

    else:
        t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

        # 독립표본일 때도 평균차이 CI를 Welch 방식으로 계산
        diff_mean = mean1 - mean2
        se = np.sqrt((std1**2)/group1.size + (std2**2)/group2.size)
        # Welch-Satterthwaite df
        df = (std1**2/group1.size + std2**2/group2.size)**2 / (
            (std1**2/group1.size)**2/(group1.size-1) + (std2**2/group2.size)**2/(group2.size-1)
        )
        t_crit = stats.t.ppf(1 - alpha/2, df=df)
        ci_low, ci_high = diff_mean - t_crit*se, diff_mean + t_crit*se

        d = _cohens_d_welch(group1, group2)

        print(f"Mean(CTMP_GIN - {baseline_name}) = {diff_mean:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{ci_low:.6f}, {ci_high:.6f}]")
        # print(f"Cohen's d (independent)         = {d:.3f}")

    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value     = {p_value:.6f}")

    '''if p_value < alpha:
        print(f"결과: p < {alpha} → 통계적으로 유의미")
    else:
        print(f"결과: p ≥ {alpha} → 통계적으로 유의미하지 않음")'''
    print()

# 실행
perform_t_test(ctmp_gin_auc, gin_auc, 'GIN', 'paired')
perform_t_test(ctmp_gin_auc, a3tgcn_auc, 'A3TGCN_2_points', 'paired')

perform_t_test(ctmp_gin_auc, gin_auc, 'GIN', 'independent')
perform_t_test(ctmp_gin_auc, a3tgcn_auc, 'A3TGCN_2_points', 'independent')
```

```python
import numpy as np
from scipy import stats

# 1. 각 모델의 K-fold 최종 Test AUC 값 입력
ctmp_gin_auc = [0.95539, 0.95488, 0.95517, 0.95493, 0.95506, 0.95365, 0.95374, 0.95369, 0.95379, 0.95378, 0.95386, 0.95431, 0.95346, 0.95443, 0.95389]
gin_auc      = [0.95339, 0.95356, 0.95344, 0.95351, 0.95322, 0.95336, 0.95325, 0.95343, 0.95312, 0.95321, 0.95361, 0.95321, 0.95367, 0.95318, 0.95368]
a3tgcn_auc   = [0.95235, 0.95210, 0.95286, 0.95268, 0.95268, 0.95223, 0.95247, 0.95241, 0.95247, 0.95285, 0.95257, 0.95310, 0.95301, 0.95296, 0.95264]

def _corrected_mean_diff_ci(diff, correction_ratio, alpha=0.05):
    """
    Nadeau and Bengio 보정이 적용된 신뢰구간 및 표준오차 계산
    """
    diff = np.asarray(diff, dtype=float)
    n = diff.size
    mean_diff = diff.mean()
    var_diff = diff.var(ddof=1) # S^2 (표본 분산)
    
    # 보정된 분산 및 표준오차(SE) 계산
    # Variance = S^2 * (1/n + n_test/n_train)
    corrected_var = var_diff * (1/n + correction_ratio)
    corrected_se = np.sqrt(corrected_var)
    
    # 신뢰구간 계산
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
    ci_low = mean_diff - t_crit * corrected_se
    ci_high = mean_diff + t_crit * corrected_se
    
    return mean_diff, corrected_se, (ci_low, ci_high)

def perform_t_test(group1, group2, baseline_name, test_type='corrected_paired', correction_ratio=0.25, alpha=0.05):
    group1 = np.asarray(group1, dtype=float)
    group2 = np.asarray(group2, dtype=float)

    mean1, std1 = group1.mean(), group1.std(ddof=1)
    mean2, std2 = group2.mean(), group2.std(ddof=1)

    print(f"--- 모델 성능 요약 ({test_type} T-test) ---")
    print(f"CTMP_GIN      : Mean = {mean1:.5f}, Std = {std1:.5f}")
    print(f"{baseline_name:<12}: Mean = {mean2:.5f}, Std = {std2:.5f}")
    print("-" * 40)

    if test_type == 'corrected_paired':
        if group1.size != group2.size:
            raise ValueError("paired t-test는 두 그룹 길이가 같아야 합니다.")
        
        diff = group1 - group2
        n = diff.size
        
        # 보정된 통계량 가져오기
        mean_diff, corrected_se, (ci_low, ci_high) = _corrected_mean_diff_ci(diff, correction_ratio, alpha)
        
        # 보정된 T-statistic 및 P-value 계산
        t_stat = mean_diff / corrected_se
        p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df=n-1))

        print(f"Mean(CTMP_GIN - {baseline_name}) = {mean_diff:.6f}")
        print(f"Corrected SE                  = {corrected_se:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{ci_low:.6f}, {ci_high:.6f}]")
        print(f"Corrected T-statistic         = {t_stat:.4f}")
        
    else:
        # 기존 일반 Paired T-test (비교용)
        t_stat, p_value = stats.ttest_rel(group1, group2)
        diff = group1 - group2
        n = diff.size
        se = diff.std(ddof=1) / np.sqrt(n)
        mean_diff = diff.mean()
        t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
        
        print(f"Mean(CTMP_GIN - {baseline_name}) = {mean_diff:.6f}")
        print(f"Standard SE                   = {se:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{mean_diff - t_crit*se:.6f}, {mean_diff + t_crit*se:.6f}]")
        print(f"Standard T-statistic          = {t_stat:.4f}")

    print(f"P-value                       = {p_value:.6f}")

    if p_value < alpha:
        print(f"결과: p < {alpha} → 통계적으로 유의미")
    else:
        print(f"결과: p ≥ {alpha} → 통계적으로 유의미하지 않음")
    print("\n")

# 실행: correction_ratio 지정 필요
# 테스트셋 비율이 15% (test=0.15) 인 경우: n_test/n_train = 0.15 / 0.85 = 0.17647
ratio = 0.15 / 0.85

print("==== 보정된 검정 (Corrected Resampled Paired T-test) ====")
perform_t_test(ctmp_gin_auc, gin_auc, 'GIN', test_type='corrected_paired', correction_ratio=ratio)
perform_t_test(ctmp_gin_auc, a3tgcn_auc, 'A3TGCN_2_pts', test_type='corrected_paired', correction_ratio=ratio)

print("==== 일반 검정 (Standard Paired T-test, 참고용) ====")
perform_t_test(ctmp_gin_auc, gin_auc, 'GIN', test_type='standard_paired')
```

```python
import numpy as np
from scipy import stats

# 1. 각 모델의 K-fold 최종 Test AUC 값 입력 (시드별로 묶인 중첩 리스트)
# 형태: [[seed1_fold1, seed1_fold2...], [seed2_fold1...], [seed3_fold1...]]
ctmp_gin_auc_raw = [
    [0.95539, 0.95488, 0.95517, 0.95493, 0.95506], 
    [0.95365, 0.95374, 0.95369, 0.95379, 0.95378], 
    [0.95386, 0.95431, 0.95346, 0.95443, 0.95389]
]
gin_auc_raw = [
    [0.95339, 0.95356, 0.95344, 0.95351, 0.95322], 
    [0.95336, 0.95325, 0.95343, 0.95312, 0.95321], 
    [0.95361, 0.95321, 0.95367, 0.95318, 0.95368]
]
# A3TGCN도 다른 모델과 동일하게 3개 시드, 5개 폴드 형식으로 변환
a3tgcn_auc_raw = [
    [0.95235, 0.95210, 0.95286, 0.95268, 0.95268],
    [0.95223, 0.95247, 0.95241, 0.95247, 0.95285],
    [0.95257, 0.95310, 0.95301, 0.95296, 0.95264]
]


def preprocess_seed_average(raw_data):
    """
    중첩 리스트 형태의 데이터를 입력받아, 동일한 Fold(열)에 대한 
    여러 시드(행)의 평균을 계산하여 1차원 배열로 반환합니다.
    """
    # np.array로 변환 후, axis=0 (열 방향, 즉 같은 fold끼리) 평균 계산
    arr = np.array(raw_data)
    averaged_folds = np.mean(arr, axis=0)
    return averaged_folds

# 전처리 실행: (3, 5) 형태의 데이터가 길이가 5인 1차원 배열(Fold 1~5의 평균)로 변환됨
ctmp_gin_auc = preprocess_seed_average(ctmp_gin_auc_raw)
gin_auc      = preprocess_seed_average(gin_auc_raw)
a3tgcn_auc   = preprocess_seed_average(a3tgcn_auc_raw)


def _mean_diff_ci(diff, alpha=0.05):
    """
    diff: array of (group1 - group2)
    returns: (mean_diff, (ci_low, ci_high))
    """
    diff = np.asarray(diff, dtype=float)
    n = diff.size
    mean_diff = diff.mean()
    sd = diff.std(ddof=1)
    se = sd / np.sqrt(n)
    t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
    ci_low = mean_diff - t_crit * se
    ci_high = mean_diff + t_crit * se
    return mean_diff, (ci_low, ci_high)

def _cohens_d_paired(diff):
    """Cohen's dz for paired samples."""
    diff = np.asarray(diff, dtype=float)
    return diff.mean() / diff.std(ddof=1)

def _cohens_d_welch(group1, group2):
    """Cohen's d for independent samples using pooled SD (classic)."""
    x = np.asarray(group1, dtype=float)
    y = np.asarray(group2, dtype=float)
    n1, n2 = x.size, y.size
    s1, s2 = x.std(ddof=1), y.std(ddof=1)
    sp = np.sqrt(((n1-1)*s1**2 + (n2-1)*s2**2) / (n1+n2-2))
    return (x.mean() - y.mean()) / sp

def perform_t_test(group1, group2, baseline_name, test_type='paired', alpha=0.05):
    group1 = np.asarray(group1, dtype=float)
    group2 = np.asarray(group2, dtype=float)

    mean1, std1 = group1.mean(), group1.std(ddof=1)
    mean2, std2 = group2.mean(), group2.std(ddof=1)

    print(f"--- 모델 성능 요약 ({test_type} T-test) ---")
    print(f"CTMP_GIN      : Mean = {mean1:.5f}, Std = {std1:.5f}")
    print(f"{baseline_name:<12}: Mean = {mean2:.5f}, Std = {std2:.5f}")
    print("-" * 40)

    if test_type == 'paired':
        if group1.size != group2.size:
            raise ValueError("paired t-test는 두 그룹 길이가 같아야 합니다.")
        t_stat, p_value = stats.ttest_rel(group1, group2)

        diff = group1 - group2
        mean_diff, (ci_low, ci_high) = _mean_diff_ci(diff, alpha=alpha)
        d = _cohens_d_paired(diff)  # dz

        print(f"Mean(CTMP_GIN - {baseline_name}) = {mean_diff:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{ci_low:.6f}, {ci_high:.6f}]")
        print(f"Cohen's dz (paired effect size) = {d:.3f}")

    else:
        t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)

        # 독립표본일 때도 평균차이 CI를 Welch 방식으로 계산
        diff_mean = mean1 - mean2
        se = np.sqrt((std1**2)/group1.size + (std2**2)/group2.size)
        # Welch-Satterthwaite df
        df = (std1**2/group1.size + std2**2/group2.size)**2 / (
            (std1**2/group1.size)**2/(group1.size-1) + (std2**2/group2.size)**2/(group2.size-1)
        )
        t_crit = stats.t.ppf(1 - alpha/2, df=df)
        ci_low, ci_high = diff_mean - t_crit*se, diff_mean + t_crit*se

        d = _cohens_d_welch(group1, group2)

        print(f"Mean(CTMP_GIN - {baseline_name}) = {diff_mean:.6f}")
        print(f"{int((1-alpha)*100)}% CI for mean diff        = [{ci_low:.6f}, {ci_high:.6f}]")
        print(f"Cohen's d (independent)         = {d:.3f}")

    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value     = {p_value:.6f}")

    if p_value < alpha:
        print(f"결과: p < {alpha} → 통계적으로 유의미")
    else:
        print(f"결과: p ≥ {alpha} → 통계적으로 유의미하지 않음")
    print("\n")


# 실행 (Paired T-test만 수행)
print("=== 전처리 적용 (시드 평균) 후 Paired T-test ===")
perform_t_test(ctmp_gin_auc, gin_auc, 'GIN', 'paired')
perform_t_test(ctmp_gin_auc, a3tgcn_auc, 'A3TGCN_2_pts', 'paired')
```
